```python
import pandas as pd
```

```python
raw_df = pd.read_csv("data/match_map_stats.csv")
```

```python
raw_df.head()
```

```python
match_ids = raw_df.match_id.unique()
len(match_ids)
```

```python
team_one_list = []
team_two_list = []
date_list = []
winner_list = []
stage_list = []

for m in match_ids:
    temp_df = raw_df[raw_df['match_id'] == m]
    team_one_list.append(temp_df.team_one_name.unique()[0])
    team_two_list.append(temp_df.team_two_name.unique()[0])
    date_list.append(temp_df.round_start_time.unique()[0])
    winner_list.append(temp_df.match_winner.unique()[0])
    stage_list.append(temp_df.stage.unique()[0])
```

```python
print(len(team_one_list))
print(len(team_two_list))
print(len(date_list))
print(len(winner_list))
print(len(stage_list))
```

```python
#Let's zip this all up and make a dataframe!
zipped_data = list(zip(match_ids, team_one_list, team_two_list, date_list, stage_list, winner_list))
columns = ['id', 'team_one', 'team_two', 'date', 'stage', 'winner']
```

```python
print(zipped_data)
```

```python
df = pd.DataFrame(zipped_data, columns=columns)
```

```python
df_temp = df[df['date'] > '01/01/2020']
```

```python
#We need to create a team list:
team_list = df.team_one.unique()
print(team_list)
```

```python
#Let's create an id list...
id_list = df_temp.id.unique()
len(id_list)
```

```python
for i in id_list:
    t_1 = ((df_temp[df_temp['id'] == i]).team_one).iloc[0]
    t_2 = ((df_temp[df_temp['id'] == i]).team_two).iloc[0]
    date = ((df_temp[df_temp['id'] == i]).date).iloc[0]
    #print(f"Team 1: {t_1} Team 2: {t_2} Date: {date}")

    team=t_1
    df_partial_season = df_temp[df_temp['date'] < date]
    #Let's get team 1 wins and losses so far....
    t_1_season_1 = df_partial_season[df_partial_season['team_one'] == team]
    t_1_season_2 = df_partial_season[df_partial_season['team_two'] == team]
    t_1_team_season = t_1_season_1.append(t_1_season_2)
    #print(f"{len(t_1_season_1)} {len(t_1_season_2)} {len(t_1_team_season)} ")
    t_1_wins = len(t_1_team_season[t_1_team_season['winner'] == team])
    t_1_losses = len(t_1_team_season[t_1_team_season['winner'] != team])
    t_1_matches = t_1_wins+t_1_losses
    #print(f"{team} {len(t_1_season_1)} {len(t_1_season_2)} {len(t_1_team_season)} {t_1_wins} ")
    #We can use a mask...
    mask = df['id'] == i
    df['t1_wins_season'][mask] = t_1_wins
    df['t1_losses_season'][mask] = t_1_losses
    df['t1_matches_season'][mask] = t_1_matches

    team=t_2
    df_partial_season = df_temp[df_temp['date'] < date]
    #Let's get team 1 wins and losses so far....
    t_1_season_1 = df_partial_season[df_partial_season['team_one'] == team]
    t_1_season_2 = df_partial_season[df_partial_season['team_two'] == team]
    t_1_team_season = t_1_season_1.append(t_1_season_2)
    #print(f"{len(t_1_season_1)} {len(t_1_season_2)} {len(t_1_team_season)} ")
    t_1_wins = len(t_1_team_season[t_1_team_season['winner'] == team])
    t_1_losses = len(t_1_team_season[t_1_team_season['winner'] != team])
    t_1_matches = t_1_wins+t_1_losses
    #print(f"{team} {len(t_1_season_1)} {len(t_1_season_2)} {len(t_1_team_season)} {t_1_wins} ")
    #We can use a mask...
    mask = df['id'] == i
    df['t2_wins_season'][mask] = t_1_wins
    df['t2_losses_season'][mask] = t_1_losses
    df['t2_matches_season'][mask] = t_1_matches
    
```
